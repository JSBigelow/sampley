## ```sampley``` tests - various miscellaneous functions

### Setup

````python
from sampley import *
import matplotlib.pyplot as plt
from math import exp
import numpy as np
import random
import typing
````

#### ```generate_dfls``` - Generate distances from line

````python
from sampley.functions import generate_dfls
````

##### With no function

###### Brief

````python
number = 100000
esw = 2000
interval = 1
dfunc = None
````

````python
dfls = generate_dfls(number=number, esw=esw, interval=interval, dfunc=dfunc)
print(dfls[:5], 'length:', len(dfls))
````

````python
plt.hist(dfls, bins=50, facecolor='none', edgecolor='black');
````

###### In-depth

````python
intervals = np.arange(0, esw + interval, interval)
intervals
````

````python
probabilities = [dfunc(x) for x in intervals] if isinstance(dfunc, typing.Callable) else np.full(len(intervals), 1)
print(probabilities[:5], 'length:', len(probabilities))
````

````python
plt.scatter(intervals, probabilities, facecolor='none', edgecolor='black');
````

````python
dfls = [float(dfl) for dfl in random.choices(intervals, probabilities, k=number)]
print(dfls[:5], 'length:', len(dfls))
````

````python
plt.hist(dfls, bins=50, facecolor='none', edgecolor='black');
````

##### With distance function

###### Brief

````python
number = 100000
esw = 2000
interval = 1
def dfunc(x): return exp((-x**2) / (2*500**2))
````

#### ```calculate_area_udf``` - Calculate area under detection function

````python
from sampley.functions import calculate_area_udf
````

##### With no function

###### Brief

````python
number = 100000
esw = 2000
interval = 1
dfunc = None
````

````python
calculate_area_udf(esw=esw, interval=interval, dfunc=dfunc)
````

###### In-depth

````python
intervals = np.arange(0, esw + interval, interval)
intervals
````

````python
probabilities = [dfunc(x) for x in intervals] if isinstance(dfunc, typing.Callable) else np.full(len(intervals), 1)
print(probabilities[:5], 'length:', len(probabilities))
````

````python
plt.scatter(intervals, probabilities, facecolor='none', edgecolor='black');
````

````python
area_udf = sum(probabilities[1:-1]*interval) + probabilities[0]*(interval/2) + probabilities[-1]*(interval/2)
area_udf
````
